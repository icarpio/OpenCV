# Flip Image

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Oct  9 21:26 2019

@author: Icarpio
"""
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
width = 500
height = 400
dim = (width, height)
# resize image
original = cv2.imread('rosalia.jpg')
resized = cv2.resize(original, dim, interpolation = cv2.INTER_AREA)
#Vertical
flipVer = cv2.flip(resized, 0)
#Horizontal
flipHor = cv2.flip(resized, 1)
#Flip Both Image
flipBoth = cv2.flip(resized, -1)

"""
Script better: 
cv2.imshow('Title', img)
cv2.imshow('Flip Both Image ', flipBoth)
cv2.waitKey(0)
cv2.destroyAllWindows()
"""
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

```python
display(resized)
```

```python
flipVer = cv2.cvtColor(flipVer, cv2.COLOR_BGR2RGB)
display(flipVer)
```

```python
flipBoth = cv2.cvtColor(flipBoth, cv2.COLOR_BGR2RGB)
display(flipBoth)
```

```python
rows, cols = resized.shape[:2]
src_points = np.float32([[0,0], [cols-1,0],[0,rows-1]])
dst_points = np.float32([[cols-1,0], [0,0],[cols-1,rows-1]])
matrix = cv2.getAffineTransform(src_points, dst_points)
mirror = cv2.warpAffine(resized, matrix, (cols,rows))
mirror = cv2.cvtColor(mirror, cv2.COLOR_BGR2RGB)
display(mirror)
```
