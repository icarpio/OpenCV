# HARRIS CORNER DETECTION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
chess = cv2.imread('chessboard.png')
chess = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(chess)
```

```python
gray_chess = cv2.cvtColor(chess, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_chess)
```

```python
plt.imshow(gray_chess, cmap='gray')
```

```python
real_chess = cv2.imread('real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray = np.float32(gray_chess)
```

```python
#cv.CornerHarris(image, harris_dst, blockSize, aperture_size=3, k=0.04) → None
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
chess[dst>0.01*dst.max()] = [255,0,0] #RGB
```

```python
plt.imshow(chess)
```

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray, blockSize =2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
real_chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(real_chess)
```

# SHI-TOMASHI DETECTION

```python
chess = cv2.imread('chessboard.png')
chess = cv2.cvtColor(chess, cv2.COLOR_BGR2RGB)
gray_chess = cv2.cvtColor(chess, cv2.COLOR_BGR2GRAY)

real_chess = cv2.imread('real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_chess,64,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for c in corners:
    x,y = c.ravel()
    cv2.circle(chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,100,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,255,-1)

plt.imshow(real_chess)
```
