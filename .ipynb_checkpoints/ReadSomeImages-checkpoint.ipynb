# Read images

```python
import os
import glob
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
images = []
filenames = []
path='C:/OpenCV/Images/*.jpg'

#Sizes 500,500
#cv2.imread(img_path)
#cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
for img_path in glob.glob(path):
    images.append(cv2.resize(cv2.cvtColor(cv2.imread(img_path),cv2.COLOR_BGR2RGB),(500,500))) #Sizes 500,500
    filenames.append(img_path)
```

```python
#Extract titles from filenames-images
names = []
titles = []
for f in filenames:
        a = f.split('/')
        names.append(a[6])
        b = a[6].split('\\')
        titles.append(b[1])
print(titles)
```

```python
fig = plt.gcf()
fig.set_size_inches(16,16)
columns = 3
for i, image in enumerate(images):
    plt.subplot(len(images) / columns + 1, columns, i + 1)
    plt.title('Image: ' + str(titles[i]))
    
    plt.imshow(image)
```
